#### STEP BY STEP 파이썬 비즈니스 통계분석

1장. 마케팅과학을 위한 통계분석

1.1 마케팅의 변화

* 데이터 분석에 기반한 마케팅 통찰력을 이끌어내는 역량의 중요성 부각 ! 
    
* 단기적인 매출증대 보다는 중장기적인 기업의 성장을 도모하고, 단순한 고객의 눈요기나 흥미 유발보다는 고객의 니즈에 부합하는 과학적인 마케팅이 이루어지기 위함 
    
* 과학적인 마케팅 중요해진 배경
    
    -   제품 중심에서 고객 중심으로의 전환
        
        세분화된 고객분류 기준 변화로 고객 중심 마케팅으로 전환을 위해 고객 데이터 분석 선행 필요
        
    -   시장점유율 확대 중심에서 고객점유율 확대 중심으로의 변화
    
        시장점유율은 자사의 총 판매량 정보만 알면 쉽게 산출 가능하지만, 고객점유율은 고객 개개인의 구매력 정보를 획득하거나, 고객들의 잠재 소비량을 통계적으로 추론해야 함
        
    -   매출액 중심에서 수익성 중심으로의 전환
    
        소요한 마케팅 비용에 따라 실질적인 수익은 천차만별, 비용 효과적인 고객반응을 이끌어 내기 위해 적중률 높은 마케팅 활동 필요
        
        고객반응률을 극대화 할 수 있는 타케팅 역량과 고객 프로파일 정보와 구매 이력을 함께 고려한 데이터 모델링 작업 필요

    -   4P에서 4R로 기반 전략
    
        Right Customer / Right Product / Right Time / Right Channel 중심의 마케팅
        체계적인 고객의 니즈 파악 / 고객-상품간의 매칭 분석 / 마케팅에 대한 고객의 최적 반응 시점과 선호 채널 검토
        
        고객으로부터 발생하는 다양한 데이터 수집 및 관리하고 분석할 수 있는 역량 필요

    -   직감에 의한 머케팅에서 근거에 의한 마케팅 기획으로 전환
    
        무차별적 실행 위주의 단기적 효과성 독특한 마케팅 활동이 아닌, 고객정보 기반의 철저한 분석과 기획, 그리고 실행 후의 피드백 과정을 통한 조직의 학습과정이 함께 동반되어야 함

    -   불특정 다수를 위한 마케팅에서 개인화된 마케팅으로 전환
        
        모든 개별 고객에 대한 마케팅 전략 수립은 불가능하지만, 최소한 개별 고객의 니즈가 모두 다르다는 것을 인지하고 세분화 가능한 범위까지 차별화된 대응전략 역량 필요





1.2 마케팅과 통계

*   통계분석기법이 마케팅 영역에 제공할 기능

    -   문제점과 기회영역의 도출

        각종 통계지표와 시각적 분석기법들은 마케터로 하여금 새로운 마케팅 활동을 필요로 하는 문제영역이나 기회영역을 발견할 수 있도록 지원
    
    -   마케팅 가설의 검정

        다양한 다변량 통계 분석 기법들은 마케터로 하여금 그들이 수립한 마케팅 가설을 객관적인 근거를 통해 검증하고, 이를 기반으로 보다 신뢰성 있는 마케팅 전략을 기획할 수 있는 기반을 만들어 줌

    -   타겟 마케팅을 위한 대상자의 추출

        통계분석기법을 활용하여 다양한 목적의 고객 점수화 모형을 개발할 수 있으며, 적중률 높은 타깃 고객층을 선별할 수 있도록 지원

    -   데이터의 요약 및 그룹화

        군집분석과 같은 통계분석기법은 단순 외형적 분류기준이 아닌 다양한 정보를 동시에 고려함으로써 유사성을 바탕으로 고객, 상품, 매장과 같은 많은 개체들을 유사한 몇 개의 집단으로 그룹핑 기능 제공 

    -   분류를 위한 의사결정

        판별분석이나 로지스틱 회귀분석과 같은 기법들은 반응고객과 미반응고객, 우수고객과 일반고객, 이탈고객과 유지고객을 분류할 수 있는 구체적인 방법론을 제공

    -   마케팅 목표치의 설정

        다중회귀분석과 같은 다변량 통계분석 기법은 다양한 원인변수를 설정하고, 이를 통해 설득력 있는 목표치를 추론할 수 있는 방법론을 제공

    -   마케팅 과정의 모니터링

        마케팅이 진행되는 과정에서도 관찰이 필요한 주요 지표들을 중심으로 지속적인 모니터링이 필요

        각종 통계량 데이터를 활용한 OLAP과 같은 시각적인 분석도구는 마케팅 현상에 대한 입체적인 분석 가능하게 함
    
    -   정확한 마케팅 성과 평가

        탄순 투자액 대비 매출액 기준 ROI(Return On Investment) 평가뿐 아니라 시계열적 패턴 대비 성과, 지역별 성과 차이의 통계적 유의도, 마케팅 적용 그룹과 비적용 그룹간의 차이 등의 성과 평가 및 통계적 검증

    -   성공과 실패의 원인 규명

        회귀분석, 분산분석과 같은 독립변수에 의한 종속변수의 변화를 설명할 수 있는 통계분석 기법으로 인과(?) 상관 관계를 통계적으로 규명할 수 있는 방법 제공

        
    

        

1.3 마케팅을 위한 통계분석 기법

*   기초 통계 분석 기법

    -   기술통계량

        평균, 분산, 표준편차, 빈도, 비율, 최빈값, 중위값 등 기본적 통계값 의미하며, 전반적인 현황 검토 및 외형적 차이 파악에 사용

        지역별, 매장별, 상품별 고객의 성별/나이 비율, 고객 방문빈도, 매출액 평균, 객단가의 계산 등 계산에 사용할 수 있음

    -   T검정

        특정 수치형 변수에 대해 두 집단간의 차이를 통계적으로 검증할 때 주로 사용

        단일표본 t-test : 예상하는 평균값이 통계적으로 유효한지 검증할 때 사용 
                         
                         (ex. 특정 브랜드의 구매 평균 연령이 32세라고 알려졌을 때, 과연 그러한지 통계적으로 검증)

        독립표본 t-test : 완전히 독립적인 두 집단의 평균 차이가 통계적으로 유의한지 판단할 때 사용 
                          
                         (ex. 맴버십 프로그램 가입 고객군과 비가입 고객군의 매출액이나 구매빈도와 같은 데이터의 차이를 통계적으로 검증)

        쌍체표본 t-test : 동일하지만 특정 사건이나 실험에 의해 두 개의 다른 시점으로 분류될 수 있는 그룹들간 평균 차이가 통계적으로 유의한지 판단할 때 사용

                         (ex. 현재 맴버십 프로그램 가입 고객들을 맴버십 가입 이전과 가입 이후로 구분하고, 매출액이나 구매 빈도의 차이를 검증)

    -   상관 관계 분석

        두 개 이상 변수들간의 상관관계 존재 여부를 통계적으로 검증할 때 사용

        변수의 형태 (명목, 서열, 수치(간격), 비율)에 따라 적용하는 상관 관계 분석 기법은 달라지나 해석은 모두 동일

        중복 활용이 불필요한 변수들을 걸러내거나 '변수그룹간의 상관성 파악하는 (정준상관분석)'에도 적용 가능

            (ex. 방문빈도와 구매금액의 상관관계, 성별과 상품 카테고리의 상관관계, 고객등급과 마케팅반응도의 상관관계 등)     

    -   범주형 데이터 분석

        명목척도로 구성된 변수들간의 상관성 또는 독립 여부 판단하는 통계적 검정방법으로 카이제곱(Chi-Square)검정이라고도 불림

            (ex. 행을 매장지역 분류로, 열을 상품유형 분류로 놓은 후 지역별 각 상품에 대한 구매 빈도를 교차표로 만들어 상관성 판단) 

    -   신뢰성 분석

        특정 변수의 측정값들이 일관된 수준으로 나타나고 있는지 판단하는 분석 기법으로, 선정한 변수들이 일관된 수준으로 측정되어야 한다는 전제를 내포함

            (ex. 고객 서베이 데이터에 대한 다변량 통계분석을 하기 위한 사전 분석 단계에서 많이 사용됨 )

*   인과(상관) 관계 분석 기법

    최소 2개 이상의 변수들을 원인(독립)변수, 결과(종속)변수로 구분하여 변수들간의 인과(상관)관계를 파악하고 검증하며, 대개 다변량(Mutivariate) 통계 분석 기법에 속함
        
    -   분산분석

        두 고객 그룹은 t-test로 검정 가능하고, 3개 이상의 집단에 대한 평균 값 차이를 검정하는 ANOVA 분산 분석이 있음.

    -   회귀분석

        하나 이상의 수치형 변수들을 독립변수로 또 다른 결과적인 수치형 변수를 종속변수로 설정해 개별 독립변수의 변화가 종속변수의 변화에 어떻게 영향을 미치는지 판단

        독립변수들의 값이 주어졌을 때 종속변수 값을 예측하거나, 독립변수와 종속변수간의 인과(상관)관계를 검증할 수 있는 통계 분석 기법

            (ex. 고객이 인지하는 상품가치/브랜드가치/관계가치가 고객만족과 고객로열티에 어떻게 영향을 미치는지, 고객과의 커뮤니케이션 횟수가 실제 매출에 영향을 미치는지 등)

    -   요인분석

        하위 구성요소를 갖는 상위 변수를 요인(factor)라고 할때 요인과 하위 구성 변수들의 구조가 적절한지 판단하는 다변량 통계 분석 기법 (신뢰성 분석 기법과 유사)

            (ex. 구매액/구매빈도/구매최근성/거래기간/교차구매지수 등 데이터를 요약하여 거래우량도와 같은 축약된 정보 추출)

    -   판별분석

        독립변수들과 이분형(or 다분형) 종속변수간의 인과(상관)관계를 판단하는 기법

            (ex. 고객의 나이/구매빈도/구매품목수로 고객의 이탈여부나 우량고객여부를 예측 (구매액 등을 예측하는 것은 회귀분석이 적절))

*   기타 다변량 통계 분석 기법

    -   군집분석

        개체의 속성들을 바탕으로 개체간의 거리를 계산한 후 가까운 개체를 동일한 집단으로 묶는 통계 분석 기법으로 분류에 대한 사전정보 없을 때 수행

            (ex. 마케팅적으로 의미 있는 고객 세분화, 상품속성들을 바탕으로 동질적인 상품군 형성, 전국 매장들을 동질적인 패턴의 매장 그룹핑 등)

    -   포지셔닝분석

        비교 대상이 되는 기업 혹은 상품간의 경쟁 구도나 시장 포지션의 형태를 포지셔닝맵과 같은 시각화도구로 파악하고, 다차원척도법과 상응분석이라는 통계 기법 사용

        다차원척도법 : 비교 대상 (브랜드, 상품, 매장 등)들간의 심리적 거리를 고객으로 응답 받아 개체들 간 경쟁 구도를 좌표 평면 상에 표현하는 방법

        상응분석 : 비교 대상 개체 (백화점브랜드 등)들을 주요 속성 관점 (교통편리성,상품다양성 등)에 비교 평가한 후 개체들 간 경쟁구도뿐만 아니라 속성들까지 좌표 평면 상에 표현해 경쟁 구도와 브랜드 선호도 함께 표현

    -   컨조인트분석

        상품/서비스와 같이 다양한 속성들의 조합으로 구성될 수 있는 개체들을 몇 가지 대안으로 만들고, 각 대안들의 고객평가결과를 수집해 상품/서비스를 구성하는 개별 속성들 간의 상대적 중요도와 효용 가치를 추정 및 높은 대안 선정하는 기법 

    -   비모수통계분석

        모수에 대한 가정을 전제로 하지 않고 모집단의 형태에 관계없이 주어진 데이터만으로 직접 확률을 계산하여 통계적 검증을 수행하는 기법      



1.4 통계분석을 활용한 문제 해결 과정

*   통계 데이터 분석을 통해 과학적이고, 객관적인 근거를 확보하여 유효한 의사결정을 내리기 위한 일종의 도구

*   통계 분석 프로세스

    -   문제파악 및 가설수립

            Ho(귀무가설) = 기업의 매출액은 지역에 상관없이 모두 같을 것 VS H1(대립(연구)가설) = 기업의 매출액은 지역에 따라 달라질 것

    -   유의수준 및 임계치설정

        가설 채택 또는 기각하기 위한 일종의 기준값 설정 (유의수준 선정 (1% 또는 5%가 일반적))

    -   분석 및 검정통계량 산출

2장. 확률통계개론

2.1 확률기초이론

*   확률
    
    -   경험 혹은 실험의 결과로 특정한 사건(Event)이나 결과가 발생할 가능성을 말함

    -   어떤 시행을 n번 반복했을 대 사상(event) A가 r번 일어나면 값 $\frac{r}{n}$을 사상 A의 상대 도수라고 하고, $n$의 수가 커질수록 이 상대도수는 일정한 값 $p$에 가까워지는데 이를 사상 A의 확률 $P(A)$라고 함

*   확률의 성질

    -   사상이 일어나지 않을 때, 확률은 0  ->  $P(∮) = 0$

    -   사상A가 일어나지 않을 확률은 1에서 사상A의 확률을 뺀 값  ->  $P(A^c) = 1 - P(A)$ 

    -   표본공간 Ω의 임의의 사상 A와 B에 대해 다음이 성립  ->  $P(A∪B) = P(A) + P(B) - P(A∩B)$ 

    -   서로 배반인 사상 A1, A2, ... An  ->  $P(A1∪A2...∪An) = P(A1) + P(A2) + ... + P(An)$

    -   사상 A가 B의 부분집합일 때 다음이 성립  ->  $A ⊂ B 이면 P(A) ≤ P(B)$

*   조건부확률

    -   사상 A가 일어났다는 조건하에서 사상 B가 일어날 확률  ->  $P(B|A) = P(A∩B) / P(A) (단, P(A) > 0)$

    -   다음 중 어느 하나를 만족하면 두 사상 A와 B는 서로 독립  ->  $P(B|A) = P(B) or P(A|B) = P(A) or P(A∩B) = P(A)P(B)$ 

*   확률변수

    통계적 실험의 모든 가능한 결과의 집합을 표본공간이라고 하며, 이 표본공간에서 정의된 실수 값의 함수

*   확률분포

    확률변수가 특정한 값을 가질 확률을 나타내는 함수

    -   이산확률분포

        확률변수가 가질 수 있는 값의 개수가 유한한 이산 확률변수가 가지는 확률분포를 의미

        -   이산확률분포의 특성
        
            a. 확률 총합 = 1  ->  $\displaystyle\sum_{i=1}^{\infty}{f(x_i)} = 1$

            b. 확률은 0에서 1의 값을 가짐  ->  $0 \leq f(x_i) \leq 1$

            c. 평균  ->  $\mu = E(X) = \displaystyle\sum{X_i}{f(X_i)}$

            d. 분산  ->  $\delta^2 = Var(X) = \displaystyle\sum{(X_i-\mu)}^2{f(X_i)}$

        -   베루누이분포 (Bernoulli distribution)

            베르누이 시행 : 시행결과가 성공과 실패 두 가지 경우만 있는 시행을 베르누이 시행이라고 하며 성공률이 $p$인 시행
            
            확률변수 X를 '성공 시 1, 실패 시 0'이라고 정의하면 $P(X=1) = p, P(X=0) = 1-p$ 가 되고 이러한 확률변수 $X$를 성공률 $p$인 베르누이 확률 변수라고 하며 그 분포를 베르누이 확률분포라고 함
        
            a. 베르누이 확률변수의 확률질량함수  ->  (성공률 p), $ P(X=x) = p^{x}(1-p)^{1-z} $, 단 x=1 또는 0

            b. 베르누이 분포의 평균과 분산  ->  평균$=p$, 분산$=p(1-p)$, 표준편차$=\sqrt{p(1-p)}$ 

        -   이항분포 (Binomial distribution)

            성공률 $p$인 베르누이 시행을 독립적으로 n번 반복 시행할 때 성공 횟수를 나타내는 확률 변수를 $X$라 하면, $X$는 $0...n$의 값을 갖는데 이러한 확률변수 X를 이항확률변수, 그 분포를 이항분포라고 함

            a. 이항확률변수의 확률질량함수 (시행횟수 $n$, 성공률 $p$)  ->  $P[X=x]=\begin{pmatrix}n\\x\end{pmatrix}p^{x}(1-p)^{n-x},(x=0,1,2,...,n)$ 단, $\begin{pmatrix}n\\x\end{pmatrix}=\frac{n!}{x!(n-x)!}$

            b. 이항분포의 평균과 분산  ->  평균$=np$, 분산$=np(1-p)$, 표준편차$=\sqrt{np(1-p)}$

            ※ 50%의 성공률을 갖는 텔레마케팅에 대한 성공과 실패는 베르누이 분포를 따르지만, 이러한 시행을 100회 시행했을 대 과연 몇 회가 성공할 것인지에 대한 확률분포는 이항분포를 따른다고 볼 수 있음
        
        -   포아송분포 (Poisson distribution)

            통상 발생 확률이 매우 적은 사건의 단위시간당 발생 횟수를 나타낼 때 사용되는 분포
            
            일정한 단위 시간에서의 평균발생횟수가 $m$이라고 할 때, 확률변수 $X$를 단위시간당 발생하는 사건의 수로 정의하면 확률변수 $X$는 평균이 $m$인 포아송 분포를 따름

            a. 포아송 분포의 확률질량함수 (평균$m$)  ->  $P[X=x] = \frac{e^{-m}m^{x}}{x!}$, $(x=0,1,2,...,n)$

            b. 포아송 분포의 평균과 분산  ->  평균$=m$, 분산$=m$, 표준편차$=\sqrt{m}$

            ※ 이항분포의 특수한 경우로서 베르누이 시행에서 시행횟수 $n$이 무한대이고, 성공확률 $P$가 0에 가까울 때의 확률분포를 의미함

    -   연속확률분포

        확률변수가 가질 수 있는 값의 개수가 무한한 연속형 확률변수가 가지는 확률분포를 의미

        확률밀도함수를 특정구간에서 적분함으로써 (즉, 특정구간에서 밀도함수의 아래부분의 면적을 구함으로써) 확률변수가 그 구간에 속할 확률이 얻어지게 됨

        -   연속확률변수의 확률밀도함수의 특성

            a. 모든 확률변수값에 대해 확률밀도함수의 값은 0이상  ->  $f(x) \geq 0$

            b. 연속확률변수 $X$가 두 수 $a$와 $b$ 사이의 값을 가질 확률은 확률밀도함수의 아래 부분 $a$와 $b$ 사이 면적과 같음  ->  $P(a \leq X \leq b) = \int\limits_a^b{f(x)dx}$
            
            c. 확률밀도함수 $f(x)$ 아래에 있는 전체 면적은 1  ->  $\int\limits_{-\infty}^{\infty}{f(x)dx}=1$

            d. 연속확률변수 $X$의 평균과 분산  ->  평균 : $\mu = E(X) = \int{xf(x)dx}$,     분산 : $\sigma^2 = Var(X) = \int{(x-\mu)^2f(x)dx}$
        
        -   정규분포 (Normal distribution)

            종모양으로 평균에 대해 좌우대칭인 분포로 평균 0, 분산 1로 표준화한 분포

            평균 = 중위수 = 최빈값

            분산(또는 표준편차)이 클수록 산포(데이터가 얼마나 퍼져 있는지 설명하는 지표로 범위,분산,표준편차,사분위수 등이 산포도의 지표로 사용)가 커서 퍼진 형태, 분산이 작을수록 평균 주위에 몰린 형태가 됨

        -   ※ 중심극한정리

            $x_1, ... x_n$ 을 평균 $\mu$, 분산 $\sigma^2$ 인 임의의 분포로부터 랜텀표본이 표본평균 $\bar{x} = \frac{\displaystyle\sum{x_i}}{n}$ 는 표본크기 $n$이 클 때 근사적으로 정규분포를 따름

            모집단의 분포가 정규분포가 아니더라도 표본수가 크면 표본평균의 분포는 정규분포로 근사함 (표본평균 $\bar{x}$의 기대값은 $\mu$, 분산은 $\frac{\sigma^2}{n}$)
        
        -   카이제곱($x^2$)분포 (Chi-Squared distribution)

            표준정규분포를 따르는 서로 독립적인 확률변수들을 가각 제곱한 다음 이들의 합을 통해 얻어지는 확률분포

            항상 양의 값만 가지고, 분포의 형태는 자유도(df : degree of freedom) n의 크기에 따라 변함

            카이제곱분포의 가법성 : $x_1, ... x_n$은 자유도 $K_1, ... K_n$인 카이제곱분포를 따르고, $x_1, ... x_n$이 서로 독립이면 $Y=\displaystyle\sum_{i=1}^{n}X_i$은 자유도 $\displaystyle\sum_{i=1}^{n}K_i$인 카이제곱분포를 따름
        
        -   t분포 (t-distribution)

            표준정규분포와 같이 평균값이 0, t분포는 카이제곱분포와 같이 자유도 n에 따라 분포의 형태가 달라짐

            모평균, 모평균의 차, 또는 회귀ㅜ계수의 추정이나 검정에 쓰이며, 주로 표본의 크기가 $n < 30$ 정도로 작을 때 이용
        
        -   F분포 (Fisher distribution)

            비대칭 연속확률분포로 오른쪽 긴 꼬리 모양의 분포 형태를 갖으며, 두개의 모집단 분산을 비교하거나 분산분석을 통해 세 개 이상의 모집단 평균 사이에 유의한 차이 존재 여부 검증 시 사용

            두개의 확률변수 $X1$, $X2$의 자유도가 각각 $d1$, $d2$이고, 서로 독립인 카이제곱분포를 따를 때 확률변수 F는 자유도 $(d1, d2)$인 F분포를 따름  ->  $F = \frac{\frac{X1}{d1}}{\frac{X2}{d2}} \sim F(d1,d2)$




2.2 통계기초이론

*   통계학(Statistics)

    관심의 대상에 대한 자료를 수집하고 정리 요약하며, 제한된 자료나 정보를 토대로 불확실한 사실에 대해 과학적 판단을 내릴 수 있는 방법을 제공하는 학문을 의미

    -   기술통계학(Descriptive Statistics)

        자료를 수집하고 정리하여 도표나 그래프를 만들거나 자료를 요약하여 대표값이나 산포도와 같은 자료의 전반적인 특성을 기술하는 방법을 다루는 학문 (ex. 평균, 분산, 표준편차 등)

    -   추론통계학(Inferential Statistics)

        자료에 내포되어 있는 정보를 분석하여 현재의 상태를 파악하거나 미래의 현상을 예측하는데 도움을 주는 통계적 추론을 다루는 분야 (ex. 모집단, 표본, 모수, 통계량 등)

*   모집단과 표본
    
    모집단(Population) : 관심의 대상이 되는 집단의 모든 개체의 관측값이나 자료값들의 집합

    표본(Sample) : 통계처리를 위해 모집단에서 실제로 뽑은 일부분의 관측값이나 자료값들의 집합

*   추정(Estimation)

    모집단에서 추출한 표본의 데이터를 이용하여 모집단의 값을 추측 및 예상하는 것을 추정이라고 함

    -   추정량의 성질

        미지의 모수 $\theta$로 대변되는 모집단으로부터 추출된 표본 $X_1, X_2, ... X_n$ 이라 할 때, 모수 $\theta$의 추정에 사용되는 통계량을 $\theta$의 추정량이라고 함

        a. 불편성(Un-biasedness) : 추정량의 기대치가 추정할 모수 $\theta$의 실제값과 같을 때, 추정량은 불편성을 가졌다고 하며, 이러한 추정량을 '불편추정량'이라고 함

        b. 유효성(Efficiency) : 모수 $\theta$를 기준으로 추정량의 분산이 작아야 한다는 원칙

        c. 일치성(Consistency) : 표본의 크기 $N$이 크면 클수록 추정량이 모수에 일치하게 되는 추정량을 '일치추정량'이라고 함

        d. 충분성(Sufficiency) : 추정량이 모수에 대해 모든 정보를 제공한다고 하면 그 추정량을 '충분추정량'이라고 함

    -   점추정(Point Estimation)

        추정하고자 하는 모수를 전정한 값이라 추측되는 단일수치로 추정하는 값

        모평균 점추정량 $\widehat \mu$ = $\overline X$ (표본평균 = 모평균 불편추정량)  
        
        모분산 점추정량 $\widehat \sigma^2$ = $S^2$ (표본분산 = 모분산 불편추정량)

        모표준편차 점추정량 $\widehat \sigma$ = $S$ (표본표준편차 = 모표준편차 불편추정량 )

    -   구간추정(Interval Estimation)

        모집단 평균이 포함될 확률구간인 신뢰구간을 설정함으로써 모수가 특정한 구간 내에 위치할 확률이 어느 정도라고 추정하는 방법

        ex. 모기업 고객 매출액평균이 95% 신뢰구간에서 50~60만원이라고 하면, '전체 고객의 매출액의 평균은 50~60만원 사이에 존재한다고 95% 확신함'을 의미

        신뢰구간(Confidence Interval) : 실제 모수가 있을 것이라고 예상되는 구간

        신뢰수준(Confidence Level) :  신뢰구간에 모수가 위치할 것이라 판단되는 믿음의 정도를 의미

    -   가설검정 (Test of Hypothesis)

        -   귀무가설 (Null Hypothesis) : 일반적으로 기존에 사실로 받아들여지고 있는 내용

        -   대립(연구)가설 (Alternative Hypothesis) : 귀무가설에 대한 반박으로서 검정을 통해 새롭게 주장하고자 하는 사실

        -   가설의 판단

            귀무가설의 유의확률(P-value)이 $\alpha$(유의수준)보다 작을 때, 귀무가설은 기각되고 연구(대립)가설 채택 

            ※ 유의확률(P-value)은 귀무가설이 참일 확률로 유의수준보다 유의확률값이 클수록 귀무가설은 기각할 수 없음   

            ![](./image/003_%EC%9C%A0%EC%9D%98%EC%88%98%EC%A4%80%EA%B3%BC%20%EA%B0%80%EC%84%A4%EC%B1%84%ED%83%9D%20%EC%98%88%EC%8B%9C.png)

        -   가설검정의 오류

            제1종오류(type I error, $\alpha$오류) : 실제 귀무가설이 옳음에도 불구하고, 귀무가설을 기각하고 대립가설을 채택하여 발생하는 오류

            제2종오류(type II error, $\beta$오류) : 실제 대립가설이 옳음에도 불구하고, 귀무가설을 기각하지 못하고 귀무가설을 채택하여 발생하는 오류

            ※ 1-$\beta$값은 대립가설에서 주장하는 모형의 검정력(Power)를 의미

        -   가설검정의 과정

            예시) 모기업 고객만족도점수가 정규분포를 따르고 평균 8.5, 표준편차 1로 알고 있다. 그러나, 최근 고객 100명 대상으로 만족도조사를 한 결과 평균 8.7을 얻음. 기존에 알려진 고객만족도가 아니라고 주장 할 수 있는지 검정. 

            step1. 가설설정

                귀무가설(H0) : 모기업의 고객만족도 평균은 8.5
                대립가설(H1) : 모기업의 고객만족도 평균은 8.5가 아님

            step2. 유의수준설정 : 보편적으로 0.05 또는 0.01로 설정함

            step3. 검정통계량결정 및 임계치 산출

                통상적으로 모평균에 대한 검정에서 사용되는 통계량은 표본평균이 되며, 모집단의 정규분포여부, 표본의 크기 및 모분산이 알려져 있는지의 여부에 따라 검정통계량의 적절한 분포가(정규분포 또는 t분포 등) 결정됨
            
            ※ 모집단이 정규분포이고 표준편차가 알려져 있는 경우, 표준정규분포를 따르는 검정통계량을 활용
            $$Z = \frac{\overline{X} - \mu_0}{\frac{\sigma}{\sqrt{n}}}$$

            ※ 모집단이 정규분포이고 표준편차가 알려지지 않은 경우, 표본분산을 사용하여 자유도 n-1을 갖는 t분포를 따르는 검정통계량을 활용
            $$T = \frac{\overline{X} - \mu_0}{\frac{s}{\sqrt{n}}}$$

            step4. 검정통계량값 계산

                검정통계량의 임계치와 비교하기 위해 표본데이터를 분석해 실제 검정통계량을 계산
            표본수 $n=100$, 모평균 $\mu_0=8.5$, 표준편차 $\sigma=1$, 표본평균 $\overline{x}=8.7$   ->  $(8.7-8.5)/(1/10)=2$

            step5. 가설 채택 여부 결정 : 본 예시에서 검정통계량값은 2로 1.96보다 크므로 귀무가설을 기각하고 연구가설을 채택. 모기업의 고객만족도평균은 8.5라고 할 수 없다라는 결론 내릴 수 있음
